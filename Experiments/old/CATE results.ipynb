```python
ihdp_flip = pd.read_csv('ihdp_rate_of_flipping.csv', header =0, names = ['name', 'rate'])
```

```python
ihdp_flip['gen'] = [i.split('_')[0] for i in ihdp_flip['name']]
```

```python
ihdp_flip['learner'] = [i[-6:] for i in ihdp_flip['name']]
```

```python
round(ihdp_flip.groupby(['gen', 'learner']).mean(), 3)
```
